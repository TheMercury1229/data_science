```python
import numpy as np
import pandas as pd
# way to plot graph
import matplotlib.pyplot as plt
import seaborn as sns
ar = np.array([4,5,6,7,18,29,30])
avg = np.sum(ar)/len(ar)
print(avg)
np.mean(ar)   #way to directly find avg using np


# Importing a random data and then calculating avg
data_set = pd.read_csv("./sample_data/california_housing_train.csv")
# avg =data_set["housing_median_age"].mean()
avg= np.mean(data_set['housing_median_age'])
print(avg)
sns.histplot(x='housing_median_age',data=data_set,bins=[i for i in range(0,81,10)])
plt.show()
```

```python
#finding median
md=np.median(data_set["housing_median_age"])
sns.histplot(x='housing_median_age',data=data_set,bins=[i for i in range(0,81,10)])
plt.plot([avg for i in range(0,17000)],[i for i in range(0,17000)],c="red")
plt.plot([md for i in range(0,17000)],[i for i in range(0,17000)],c="blue")
plt.show()
```

```python
mode=data_set["housing_median_age"].mode()[0]
# data_set["housing_median_age"].value_counts() Gives list of all the data with frequency
sns.histplot(x='housing_median_age',data=data_set,bins=[i for i in range(0,81,10)])
plt.plot([avg for i in range(0,17000)],[i for i in range(0,17000)],c="red",label='mean')
plt.plot([md for i in range(0,17000)],[i for i in range(0,17000)],c="blue",label='median')
plt.plot([mode for i in range(0,17000)],[i for i in range(0,17000)],c="green",label='mode')
plt.show()
```
